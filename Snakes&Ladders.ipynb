# Snake and Ladder

```python
#imports
import random
```

```python
#global variables
player1 = {"name": "", "position": 0}
player2 = {"name": "", "position": 0}

player = {"player1": player1, "player2": player2}

snake = {17 : 7, 54 : 34, 62 : 19, 98 : 79}
ladder = {3 : 38, 24 : 33, 42 : 93, 72 : 84}
```

```python
#function for changing position
def nextStep(player_value, roll):
    if player[player_value]['position'] + roll <= 100:
        if snake.get(player[player_value]['position'] + roll, None) != None:
            print('snake bite!')
            player[player_value]['position'] = snake[player[player_value]['position'] + roll]
        elif ladder.get(player[player_value]['position'] + roll, None) != None:
            print('ladder, woohoo!!')
            player[player_value]['position'] = ladder[player[player_value]['position'] + roll]
        else:
            player[player_value]['position'] = player[player_value]['position'] + roll
```

```python
#auto mode gameplay
def AutoMode():
    print ("You've Started the game in Auto Mode")
    
    while True:
        play = input('Player 1 : ')
        if play.lower() == 'roll':
            roll = random.randint(1,6)
            print ("You've got a :", roll)
            nextStep('player1', roll)
            print ('Your final position is ', player1['position'])
            
            if player1['position'] == 100:
                print ("Player 1 won the game\n##### Game Successfully Finished ####")
                break
                    
        elif play.lower() == 'quit':
            print ("Player 1 Quit\nPlayer 2 won the game\n##### Game Over ####")
            break
        else:
            print ("Wrong input given, Give \"Roll\" or \"Quit\" as input")
            continue
        
        play = input('Player 2 : ')
        if play.lower() == 'roll':
            roll = random.randint(1,6)
            print ("You've got a :", roll)
            nextStep('player2', roll)
            print ('Your final position is ', player2['position'])
            
            if player2['position'] == 100:
                print ("Player 2 won the game\n##### Game Successfully Finished ####")
                break
                    
        elif play.lower() == 'quit':
            print ("Player 2 Quit\nPlayer 1 won the game\n##### Game Over ####")
            break
        else:
            print ("Wrong input given, Give \"Roll\" or \"Quit\" as input")
    
    return
```

```python
#taking manual input
def manInput(player_value):
    while True:
        try:
            print("")
            print (player_value[:6], player_value[6:], ":")
            roll = input ("Enter an integer between 1 to 20 : or 'Quit' ")
            
            if roll.lower() == 'quit':
                print (player_value[:6], player_value[6:],"Quit")
                if player_value[6:] == '1':
                    print ("Player 2 won the game\n##### Game Over #### ")
                else:
                    print("Player 1 won the game\n##### Game Over #### ")
                    
                return True
                
            roll = int(roll)
            
            if roll >= 1 and roll <= 20:
                print ("You've got a :", roll)
                nextStep(player_value, roll)
                print ('Your final position is', player[player_value]['position'])

                if player[player_value]['position'] == 100:
                    if player_value[6:] == '1':
                        print ("Player 1 won the game\n##### Game Successfully Finished #### ")
                    else:
                        print("Player 2 won the game\n##### Game Successfully Finished #### ")
                        
                    return True
                
                return False
            
            else:
                print ("Enter a number between 1 - 20")
        except:
            print ("Enter a number between 1 - 20")
```

```python
#manual mode game play
def ManualMode():
    print ("You've Started the game in Manual Mode")
    
    while True:
        if manInput('player1'):
            break
        
        if manInput('player2'):
            break
    
    return
```

```python
#creating a switch statement in python
switcher={
        1: ManualMode,
        2: AutoMode
    }

def switch(choice):
    x = switcher.get(choice,"Incorrect input")
    
    return x
```

```python
#taking user inputs and starting the game
print("#### Welcome to Snake and Ladder game ####")
while True:
    if len(player1['name']) < 1:
        player1['name'] = input('Enter the name of player 1 : ')
    
    if len(player1['name']) < 1:
        print("Enter a valid Username")
        continue
        
    player2['name'] = input("Enter the name of Player 2 : ")
    
    if len(player2['name']) < 1:
        print("Enter a valid Username")
    else:
        break

print("### Let us start! ###")
while True:
    try:
        x = int(input('Enter a number 1 or 2 for Manual or Auto mode respectively '))
        x = switch(x)
        if not x == "Incorrect input":
            x()
            break
        else:
            print("Incorrect Input")
    except:
        print("Enter a valid input")
```
